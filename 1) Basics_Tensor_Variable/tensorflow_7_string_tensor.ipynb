```python
import tensorflow as tf
```

```python
string_tensor=tf.constant(["hellow","world","moon","gas"])
print(string_tensor)
```

```python
# tf.string.join - to join the elenmtns of the list using a spcific seperator 
print(tf.strings.join(string_tensor,separator=""))
print(tf.strings.join(string_tensor,separator="#"))
```

# tensor variables

```python
#in above image the value of a1,a2 b1,b2 gets updated as we train the model so we need some variables to hold those values, for this 
#we use tensor variables

#to define a variable
x=tf.constant([1,2])
a1=tf.Variable(x,name="var1")
print(a1)
#for above a1 act as a variable 

#to subtract [3,4] from variable a1 which holds : [1,2], we can do it as follow 
print(f"the value after subtraction is: \n {a1.assign_sub([3,4])}")

#to add [3,4] from variable a1 we use bellow line
print(f"the value after addition is : \n {a1.assign_add([3,4])}")
```

```python
#IMPORTANT---
#to defince a variable in out cpu,gpu,tpu we use the following code:
with tf.device('GPU:0'):
    a11=tf.constant(0.2)
print(f"the value of a1 is {a11}")
print(f"the device for computing a1 is {a11.device}")
```
